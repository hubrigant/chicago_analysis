```python
community_df = pd.read_pickle('data/community_modeled.pkl')
census_summary_df = pd.read_pickle('data/census_summary-cleaned.pkl')
```

```python
summary_transformed = pd.merge(census_summary_df, community_df,
                               left_on='GeogKey', right_on='community number')
```

```python
del summary_transformed['community name']
del summary_transformed['community number']
```

```python
summary_transformed['population density'] = summary_transformed['Total Population'] / summary_transformed['community area']
summary_transformed['housing density'] = summary_transformed['Total Housing Units'] / summary_transformed['community area']
summary_transformed['vacated density'] = summary_transformed['Vacant Housing Units'] / summary_transformed['community area']
summary_transformed['vacated ratio'] = summary_transformed['Vacant Housing Units'] / summary_transformed['Total Housing Units']
```
